```python
from matplotlib import *
from pylab import *
from numpy import *
from scipy import *
from scipy import stats

# --- Code English ---

X = stats.poisson(3.5)

n = arange(0,15)
fig, axes = subplots(3,1, sharex=True)

axes[0].step(n, X.pmf(n))

axes[1].step(n, X.cdf(n))

axes[2].hist(X.rvs(size=1000));

show()

# Create a (continous) random variable with normal distribution
y = stats.norm()
x = linspace( -5,5,100)
fig, axes = subplots(3,1, sharex=True)
# plot the probability distribution function (PDF)
axes[0].plot(x, y.pdf(x))
# plot the commulative distribution function (PDF)
axes[1].plot(x, y.cdf(x));
# plot histogram of 1000 random realizations of the stochastic variable Y
axes[2].hist(y.rvs(size=1000), bins =50);

show()

x.mean(), x.std(), x.var()
y.mean(), y.std(), x.var()

t_statistic, p_value = stats.ttest_ind(X.rvs(size=1000), X.rvs(size=1000))
print("t-statistic =", t_statistic)
print("p-value", p_value)


```

```python
from matplotlib import *
from pylab import *
from numpy import *
from scipy import *
from scipy import stats
## from scipy.integrate import odent, ode
from scipy.interpolate import *

def f(x):
  return sin(x)
n = arange(0,10)
x = linspace(0, 9, 100)
y_means = f(n) + 0.1 * randn(len(n)) # simulate measurement with noise
y_real = f(x)

# Linear
linear_interp = interp1d(n, y_means)
y_interp1 = linear_interp(x)

# Cubic
cubic_interp = interp1d(n, y_means, kind='cubic')
y_interp2 = cubic_interp(x)

fig, ax = subplots(figsize=(10,4))
ax.plot(n, y_means, 'bs', label='data')
ax.plot(x, y_real, 'k', lw=2, label='real')
ax.plot(x, y_interp1, 'r', lw=2, label='linear')
ax.plot(x, y_interp2, 'g', lw=2, label='cubic')
ax.legend(loc=3)
show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Dados das médias de gols
anos = [
    1930, 1934, 1938, 1950, 1954, 1958, 1962, 1966, 1970, 1974,
    1978, 1982, 1986, 1990, 1994, 1998, 2002, 2006, 2010, 2014,
    2018, 2022
]
medias_gols = [
    3.89, 4.12, 4.67, 4.00, 5.38, 3.60, 2.78, 2.78, 2.97, 2.55,
    2.68, 2.81, 2.54, 2.21, 2.71, 2.67, 2.52, 2.30, 2.27, 2.67,
    2.64, 2.69
]

# Criar um DataFrame para os dados
df = pd.DataFrame({
    'Ano': anos,
    'Média de Gols': medias_gols
})

# Configurar o gráfico
plt.figure(figsize=(14, 8))
plt.plot(df['Ano'], df['Média de Gols'], marker='o', linestyle='-', color='b')

# Adicionar títulos e rótulos
plt.title('Média de Gols por Jogo em Cada Edição da Copa do Mundo')
plt.xlabel('Ano')
plt.ylabel('Média de Gols por Jogo')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()

# Mostrar o gráfico
plt.show()
```
